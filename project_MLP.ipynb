```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo
from tabulate import tabulate
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
from tensorflow import keras
from tensorflow.keras import layers
import warnings
from sklearn.exceptions import UndefinedMetricWarning
import time

warnings.filterwarnings("ignore", category=UndefinedMetricWarning)

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
random.seed(RANDOM_STATE)

covertype = fetch_ucirepo(id=31)
X = covertype.data.features
y = covertype.data.targets
df = pd.concat([X, y], axis=1)

table = []
for i, col in enumerate(df.columns, start=1):
    role = "Target" if col == "Cover_Type" else "Feature"
    dtype = str(df[col].dtype)
    missing = "yes" if df[col].isnull().any() else "no"
    if df[col].dtype == 'object':
        value_range = f"Categories: {list(df[col].unique())}"
    else:
        nunique = df[col].nunique()
        if nunique <= 10:
            value_range = f"Values: {sorted(df[col].unique())}"
        else:
            value_range = f"Range: {df[col].min()} to {df[col].max()}"
    row = [i, col, role, dtype, "N/A", "", missing, value_range]
    table.append(row)
headers = ["No.", "Variable Name", "Role", "Type", "Description", "Units", "Missing Values", "Value Range"]
print(tabulate(table, headers=headers, tablefmt="grid"))

df = df.dropna(subset=["Cover_Type"])
X = df.drop(columns="Cover_Type")
y = df["Cover_Type"]

class_counts = y.value_counts().sort_index()
n_classes = class_counts.shape[0]
total_samples = class_counts.sum()
print("="*60)
print(f"Total number of samples: {total_samples}")
print(f"Number of target classes (Cover_Type): {n_classes}")
print("Number of samples in each target class:")
for cls, count in class_counts.items():
    print(f"Class {cls}: {count} samples")
print("="*60)

plt.figure(figsize=(10, 6))
sns.barplot(
    x=class_counts.index.astype(str),
    y=class_counts.values,
    hue=class_counts.index.astype(str),
    palette="rocket",
    legend=False
)
plt.title("Distribution of Cover_Type Classes (All Data)")
plt.xlabel("Cover_Type Class")
plt.ylabel("Number of Samples")
for idx, val in enumerate(class_counts.values):
    percent = val / total_samples * 100
    plt.text(idx, val + total_samples*0.005, f"{val} ({percent:.1f}%)", ha='center', fontsize=12)
plt.show()

max_count = class_counts.max()
min_count = class_counts.min()
imbalance_ratio = max_count / min_count
print(f"Largest class sample count: {max_count}")
print(f"Smallest class sample count: {min_count}")
print(f"Imbalance ratio (max/min): {imbalance_ratio:.2f}")
if imbalance_ratio > 3:
    print("Warning: The dataset is imbalanced!")
else:
    print("The dataset is relatively balanced.")
print("="*60)

X_50, _, y_50, _ = train_test_split(X, y, train_size=0.5, random_state=RANDOM_STATE, stratify=y)
print(f"\n50% of the data was randomly selected (with class ratio preserved):\nX_50 shape = {X_50.shape}, y_50 shape = {y_50.shape}")

print("\nTarget class distribution in all data:")
for cls, count in class_counts.items():
    percent = count / total_samples * 100
    print(f"Class {cls}: {count} samples ({percent:.2f}%)")

class_counts_50 = pd.Series(y_50.values.ravel()).value_counts().sort_index()
total_50 = class_counts_50.sum()
print("\nTarget class distribution in the 50% selected data:")
for cls, count in class_counts_50.items():
    percent = count / total_50 * 100
    print(f"Class {cls}: {count} samples ({percent:.2f}%)")

scaler = StandardScaler()
X_50_scaled = scaler.fit_transform(X_50)

X_temp, X_test, y_temp, y_test = train_test_split(X_50_scaled, y_50, test_size=0.15, random_state=RANDOM_STATE, stratify=y_50)
train_size = 70 / 85
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=(1 - train_size), random_state=RANDOM_STATE, stratify=y_temp)

print("\nDataset sizes after split:")
print(f"Training set size: {X_train.shape[0]}")
print(f"Validation set size: {X_val.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")
print("="*60)

def fix_labels(y):
    y = np.array(y).ravel()
    return (y - 1).astype(int)
y_train = fix_labels(y_train)
y_val = fix_labels(y_val)
y_test = fix_labels(y_test)

n_classes = len(np.unique(y_train))
input_dim = X_train.shape[1]
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=np.ravel(y_train))
class_weight_dict = dict(zip(np.unique(y_train), class_weights))

param_space = {
    "hidden_units": [[64, 32], [128, 64, 32], [256, 128, 64]],
    "activation": ['relu', 'tanh', 'sigmoid'],
    "optimizer": ['adam', 'sgd', 'rmsprop'],
    "learning_rate": [0.001, 0.0005, 0.005, 0.01],
    "batch_size": [128, 256],
    "loss": ['sparse_categorical_crossentropy']
}

def random_params(space):
    return {
        "hidden_units": random.choice(space["hidden_units"]),
        "activation": random.choice(space["activation"]),
        "optimizer": random.choice(space["optimizer"]),
        "learning_rate": random.choice(space["learning_rate"]),
        "batch_size": random.choice(space["batch_size"]),
        "loss": random.choice(space["loss"])
    }

def build_mlp(input_dim, n_classes, hidden_units, activation):
    model = keras.Sequential()
    model.add(layers.Input(shape=(input_dim,)))
    for units in hidden_units:
        model.add(layers.Dense(units, activation=activation))
    model.add(layers.Dense(n_classes, activation='softmax'))
    return model

N_MODELS = 3
model_archs = []
print("Selected architectures for 3 random models:\n")
for i in range(N_MODELS):
    arch = random_params(param_space)
    model_archs.append(arch)
    print(f"Model #{i+1} architecture:")
    for k, v in arch.items():
        print(f"  {k}: {v}")
    print("-" * 40)

N_REPEATS = 10
all_results = []

for idx, arch in enumerate(model_archs):
    print(f"\n{'='*60}")
    print(f'Random Model #{idx+1} Hyperparameters:')
    for k, v in arch.items():
        print(f"- {k}: {v}")
    print(f"{'='*60}")

    for mode, mode_name, class_weight in [
        ('no_balance', '[No class balancing]', None),
        ('class_weight', '[Class Weight balancing]', class_weight_dict)
    ]:
        metrics = []
        confmats = []
        print(f"\n{mode_name}")
        for i in range(N_REPEATS):
            print(f"  Repetition {i+1}/{N_REPEATS}...", end=' ')
            model = build_mlp(
                input_dim, n_classes,
                hidden_units=arch["hidden_units"],
                activation=arch["activation"]
            )
            optimizer = keras.optimizers.get({
                'class_name': arch["optimizer"],
                'config': {'learning_rate': arch["learning_rate"]}
            })
            model.compile(
                optimizer=optimizer,
                loss=arch["loss"],
                metrics=['accuracy']
            )
            
            callbacks = [keras.callbacks.EarlyStopping(patience=8, restore_best_weights=True, monitor='val_loss')]
            y_train_fit = y_train
            y_val_fit = y_val
            y_test_eval = y_test
            
            model.fit(
                X_train, y_train_fit, 
                validation_data=(X_val, y_val_fit),
                epochs=50,
                batch_size=arch["batch_size"],
                callbacks=callbacks,
                verbose=0,
                class_weight=class_weight
            )
            y_pred = model.predict(X_test, verbose=0).argmax(axis=1)
            y_true = y_test

            report = classification_report(y_true, y_pred, digits=4, output_dict=True, zero_division=0)
            acc = report["accuracy"]
            precision = report["macro avg"]["precision"]
            recall = report["macro avg"]["recall"]
            f1 = report["macro avg"]["f1-score"]
            metrics.append([precision, recall, f1, acc])
            confmats.append(confusion_matrix(y_true, y_pred))
            print(f"Precision: {precision:.4f} | Recall: {recall:.4f} | F1: {f1:.4f} | Acc: {acc:.4f}")
        
        arr = np.array(metrics)
        mean = np.round(np.mean(arr, axis=0), 2)
        std = np.round(np.std(arr, axis=0), 2)
        mean_confmat = np.round(np.mean(confmats, axis=0), 2)
        
        row_labels = ['Mean', 'Std']
        col_labels = ['Precision', 'Recall', 'F1-score', 'Accuracy']
        summary_df = pd.DataFrame([mean, std], columns=col_labels, index=row_labels)
        print(f"\nSummary Table ({mode_name}):")
        print(summary_df)
        
        print(f"\nMean Confusion Matrix ({mode_name}):")
        print(mean_confmat.astype(int))
        ConfusionMatrixDisplay(mean_confmat).plot(xticks_rotation='vertical')
        plt.title(f"Mean Confusion Matrix - Model {idx+1} - {mode_name}")
        plt.show()
        all_results.append({
            'Model': f'Model {idx+1}',
            'Mode': mode_name,
            **arch,
            'Precision (mean)': mean[0],
            'Recall (mean)': mean[1],
            'F1-score (mean)': mean[2],
            'Accuracy (mean)': mean[3],
            'Precision (std)': std[0],
            'Recall (std)': std[1],
            'F1-score (std)': std[2],
            'Accuracy (std)': std[3]
        })

final_table = pd.DataFrame(all_results)
print('\nFinal Comparison Table:')
print(final_table)

idx_best = final_table['Accuracy (mean)'].idxmax()
best_model = final_table.loc[idx_best]
print('\nBest Model by Accuracy:')
print(best_model)



def permutation_importance_numeric(
    model, X, y, metric_fn=accuracy_score, n_repeats=3, feature_names=None, random_state=42
):
    np.random.seed(random_state)
    start_time = time.time()
    if isinstance(X, pd.DataFrame):
        numeric_X = X.select_dtypes(include=[np.number]).copy()
        num_feature_names = numeric_X.columns.tolist()
        X_numeric = numeric_X.values
    else:
        num_feature_names = feature_names if feature_names is not None else [f"feature_{i}" for i in range(X.shape[1])]
        X_numeric = X

    print(f"Number of numeric features: {len(num_feature_names)}")
    print(f"Number of test samples: {X_numeric.shape[0]}")
    print("Evaluating model on original data ...")
    base_pred = model.predict(X_numeric).argmax(axis=1)
    base_score = metric_fn(y, base_pred)
    print(f"> Baseline accuracy: {base_score:.4f}")

    importances = []
    for col in range(X_numeric.shape[1]):
        feature_time = time.time()
        print(f"\nFeature [{col+1}/{X_numeric.shape[1]}]: {num_feature_names[col]}")
        scores = []
       
        for rep in range(n_repeats):
            print(f"  Repeat {rep+1}/{n_repeats} ...", end=" ")
            X_perm = X_numeric.copy()
            np.random.shuffle(X_perm[:, col])
            perm_pred = model.predict(X_perm).argmax(axis=1)
            perm_score = metric_fn(y, perm_pred)
            drop = base_score - perm_score
            scores.append(drop)
            print(f"accuracy drop: {drop:.4f}")
        mean_importance = np.mean(scores)
        print(f"Mean importance: {mean_importance:.4f} (time: {time.time()-feature_time:.1f}s)")
        importances.append(mean_importance)
    print(f"\nTotal runtime: {time.time() - start_time:.1f}s")
   
    df_importance = pd.DataFrame({"feature": num_feature_names, "importance": importances})
    df_importance = df_importance.sort_values("importance", ascending=False).reset_index(drop=True)
    return df_importance, base_score

numeric_feature_names = [f"feature_{i}" for i in range(X_test.shape[1])]
df_importance, base_acc = permutation_importance_numeric(
    model,
    X_test,
    y_test,
    metric_fn=accuracy_score,
    n_repeats=3,
    feature_names=numeric_feature_names
)
print("\n============================")
print("Permutation Feature Importances:")
print(df_importance)
plt.figure(figsize=(12,5))
bars = plt.bar(df_importance["feature"], df_importance["importance"], color='royalblue', edgecolor='black')
plt.xticks(rotation=90, fontsize=10)
plt.ylabel("Accuracy Drop After Permutation", fontsize=12)
plt.title("Permutation Importance of Numeric Features (MLP)", fontsize=13, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
for bar, value in zip(bars, df_importance["importance"]):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{value:.3f}', ha='center', va='bottom', fontsize=9, fontweight='bold', color='black')
plt.show()
```
